```python
# Line Plot (trend over time or sequence)

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 8, 20, 12]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Bar Plot (comparing categories)

x = [1, 2, 3, 4, 5]
y = [10, 15, 8, 20, 12]

plt.bar(x, y)
plt.title("Simple Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# Histogram (distribution of data)

import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(100)

plt.hist(data)
plt.title("Simple Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Countplot (frequency of categories) - Used for categorical data:

import seaborn as sns
import matplotlib.pyplot as plt

dept = ['CSE', 'ECE', 'CSE', 'EEE', 'ECE', 'CSE']

sns.countplot(x=dept)
plt.title("Countplot Example")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

```python
# Histplot (Histogram using Seaborn) - Histogram = how many values fall in each bin (range)

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(100)

sns.histplot(data, bins=10)
plt.title("Histplot Example")
plt.show()
```

```python
# Theme Palettes and Styles (Seaborn)

sns.set_style("darkgrid") 
sns.set_palette("Set2")

sns.countplot(x=dept)
plt.title("Countplot with Theme")
plt.show()
```

```python
# with titanic dataset

import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')  
```

```python
# box plot for the above dataset

plt.figure(figsize=(8, 6))
sns.boxplot(x="pclass", y="age", data=titanic, palette="Set2")
plt.title("Age Distribution by Passenger Class")
plt.show()
```

```python
# count plot

plt.figure(figsize=(6, 5))
sns.countplot(x="pclass", data=titanic, palette="pastel")
plt.title("Number of Passengers by Class")
plt.show()
```

```python
'''
Decorator Functions
A decorator is a function that:

 takes another function
 adds extra features to it
 returns a new function

uses - Logging ,Authentication, Timing function execution,Reusing common functionality
'''
```

```python
def my_decorator(func):
    def wrapper():
        print("Before the function runs...")
        func()
        print("After the function runs...")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

```python
# function Composition - Function composition = combining functions together - Output of one function becomes input of another function. - 

def f(x):
    return x + 2

def g(x):
    return x * 3

# composition f(g(x))
print(f(g(5)))   
```

```python
# difference
# Composition means combining functions where the output of one becomes the input of another.
# A decorator adds extra behavior to an existing function without modifying it. It wraps a function inside another function.
```
