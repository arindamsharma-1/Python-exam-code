```python
# handling missing data

import numpy as np

data = {
    "Student": ["Tilak", "Sarojini", "Bhagath", "Jhansi", "Vallabhai"],
    "Department": ["CSE", "ECE", "EEE", np.nan, "CSE"],
    "Marks": [85, np.nan, 78, 92, np.nan],
    "Attendance": [90, 85, np.nan, 95, 80]
}

df_students = pd.DataFrame(data)
print(df_students)



```

```python
# detecting missing values

df_students.isna()
```

```python
df_students.isna().sum() # NaNs per column count
```

```python
# Basic fillna() Usage - Replace all missing values with a single value

df1=df_students.fillna(0)
print(df1)
#print(df_students)
```

```python
import numpy as np

data = {
    "Student": ["Tilak", "Sarojini", "Bhagath", "Jhansi", "Vallabhai"],
    "Department": ["CSE", "ECE", "EEE", np.nan, "CSE"],
    "Marks": [85, np.nan, 78, 92, np.nan],
    "Attendance": [90, 85, np.nan, 95, 80]
}

df_students = pd.DataFrame(data)
print(df_students)
```

```python
# Fill missing values only in “Marks” column

df2 = df_students.copy()
df2["Marks"] = df2["Marks"].fillna(df2["Marks"].mean())
print(df2)
```

```python
df3 = df_students.copy()
df3["Department"] = df3["Department"].fillna("Unknown")
print(df3)

```

```python
# forward interpolation
df5 = df_students.copy()
df5 = df5.fillna(method="ffill")
print(df5)
print()
print(df_students)
```

```python
# backward interpolation
df6 = df_students.copy()
df6 = df6.fillna(method="bfill")
print(df6)
print()
print(df_students)
```

```python
df_drop_any = df_students.dropna() # without any missing values
print(df_drop_any)
```

```python
df_drop_all = df_students.dropna(how="all") # drops rows/cols only if all the values are null
print(df_drop_all)
```

```python
df_drop_marks = df_students.dropna(subset=["Marks"]) # dropping rows which has missing values
print(df_drop_marks)
```

```python
# interpolation

df_inter = df_students.copy()
df_inter["Marks"] = df_inter["Marks"].interpolate()
df_inter["Attendance"] = df_inter["Attendance"].interpolate()
print(df_inter)
```

```python
# grouping - marks column filled with mean as aggreagation has to be done

df = df_students.copy()
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())  # mean = 85
print(df)
```

```python
dept_mean = df.groupby("Department")["Marks"].mean()
print(dept_mean)
```

```python
dept_agg = df.groupby("Department")["Marks"].agg(["mean", "min", "max", "count"])
print(dept_agg)
```

```python
# transform() — Keeps Original Row Size -Subtract each student’s marks from their department mean

df["Marks_diff"] = df["Marks"] - df.groupby("Department")["Marks"].transform("mean")
print(df)
```
