```python
# vstack, hstack and concatenation (works like hstack)
# In vstack columns should match (axis=0) - number of cols should be same
# In hstack rows should match (axis=1) - number of rows should be same

a = np.array([[1, 2], [3, 4]])
b = np.array([[5,6]])

result = np.vstack((a, b)) # row where axis=0
print(result)
```

```python
# hstack rows should match
a = np.array([[1, 2], [3, 4]])
b = np.array([[5],[6]])


result = np.hstack((a, b)) # column where axis=1
print(result)
```

```python
# concatenate

np.concatenate((a,b),axis=0/1)
```
